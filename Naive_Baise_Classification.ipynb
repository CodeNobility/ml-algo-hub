```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
import numpy as np
```

```python
dataset=pd.read_csv(r"/Users/princegupta/Documents/data science/dataset/synthetic_placement_data.csv")
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.kdeplot(data=dataset["score"])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
x_train_transformed=ft.fit_transform(x_train)
```

```python
x_test_transformed=ft.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

# GaussianNB

```python
gb=GaussianNB()
gb.fit(x_train_transformed,y_train)
```

```python
gb.score(x_train_transformed,y_train)*100, gb.score(x_test_transformed,y_test)*100
```

```python
plt.figure(figsize=(8,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gb)
plt.show()
```

# MultinomialNB

```python
mb=MultinomialNB()
mb.fit(x_train_transformed,y_train)
```

```python
mb.score(x_train_transformed,y_train)*100, mb.score(x_test_transformed,y_test)*100
```

```python
plt.figure(figsize=(8,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mb)
plt.show()
```

# BernoulliNB

```python
bl=BernoulliNB()
bl.fit(x_train_transformed,y_train)
```

```python
bl.score(x_train_transformed,y_train)*100, bl.score(x_test_transformed,y_test)*100
```

```python
plt.figure(figsize=(8,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bl)
plt.show()
```
