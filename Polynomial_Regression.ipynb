```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"/Users/princegupta/Documents/data science/dataset/polynomial_salary_data_updated.csv")
```

```python
dataset.head(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="Level",y="Salary",data=dataset)
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
x = dataset[["Level"]]
y = dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
```

```python
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
pred = lr.predict(x)
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="Level",y="Salary",data=dataset)
plt.plot(dataset["Level"],pred,c="r")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(["org","predicted"])
plt.show()
```
