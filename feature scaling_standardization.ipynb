```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("dataset/loan_data.csv")
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["days.with.cr.line"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["days.with.cr.line"]])
```

```python
dataset["days.with.cr.line_ss"]=pd.DataFrame(ss.transform(dataset[["days.with.cr.line"]]),columns=["x"])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(dataset["days.with.cr.line"])
plt.subplot(1,2,2)
sns.distplot(dataset["days.with.cr.line_ss"])
plt.show()
```
