```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("dataset/loan_data.csv")
```

```python
sns.distplot(dataset["revol.bal"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["revol.bal"]])
```

```python
dataset["revol.bal_ft"]=ft.transform(dataset[["revol.bal"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(dataset["revol.bal"])
plt.title("before")

plt.subplot(1,2,2)
sns.distplot(dataset["revol.bal_ft"])
plt.title("after")
plt.show()
```
