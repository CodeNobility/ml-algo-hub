```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.2)
```

```python
df={"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
df
```

```python
dataset=pd.DataFrame(df)
```

```python
dataset.head(4)
```

```python
x_a=dataset.iloc[:,:-1]
y_a=dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_a,y_a,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
bg=BaggingClassifier(estimator=KNeighborsClassifier(),n_estimators=50)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train)*100,bg.score(x_test,y_test)*100
```

```python
x_test[20:25]
```

```python
y_test[20:25]
```

```python
data=pd.DataFrame([[-0.185574,0.296769]],columns=x_a.columns)
bg.predict(data)
```

```python
data1=pd.DataFrame([[-0.093101,0.971214]],columns=x_a.columns)
bg.predict(data1)
```

```python
rf=RandomForestClassifier(n_estimators=30)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)*100,rf.score(x_test,y_test)*100
```

```python
for i in range(10,80):
    rf1=RandomForestClassifier(n_estimators=i)
    rf1.fit(x_train,y_train)
    print(i,rf1.score(x_train,y_train)*100,rf1.score(x_test,y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf=confusion_matrix(y_test,bg.predict(x_test))
cf
```

```python
sns.heatmap(cf,annot=True)
plt.show()
```

```python
cf1=confusion_matrix(y_test,rf.predict(x_test))
cf1
```

```python
sns.heatmap(cf1,annot=True)
plt.show()
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
precision_score(y_test,bg.predict(x_test))
```

```python
precision_score(y_test,rf.predict(x_test))
```

```python
recall_score(y_test,bg.predict(x_test))
```

```python
recall_score(y_test,rf.predict(x_test))
```
