```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("dataset/loan_data.csv")
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["revol.util"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()
mms.fit(dataset[["revol.util"]])
```

```python
dataset["revol.util_MinMax"]=pd.DataFrame(mms.transform(dataset[["revol.util"]]),columns=["x"])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(dataset["revol.util"])
plt.subplot(1,2,2)
sns.distplot(dataset["revol.util_MinMax"])
plt.show()
```
