```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"/Users/princegupta/Documents/data science/dataset/Groceries_dataset.csv")
```

```python
dataset.head(3)
```

```python
dataset=dataset[["itemDescription"]]
```

```python
dataset.head(2)
```

```python
transaction=dataset.groupby(dataset.index)["itemDescription"].apply(list).tolist()
```

```python
transaction
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr=TransactionEncoder()
tr.fit(transaction)
```

```python
tr.transform(transaction)
```

```python
df=pd.DataFrame(tr.transform(transaction),columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(df,min_support=0.03,use_colnames=True)
```
