```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset=pd.read_csv("dataset/pima indian daibetes.csv")
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Outcome"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
sf=SequentialFeatureSelector(lr,k_features=8,forward=True)
sf.fit(x,y)
```

```python
sf.feature_names
```

```python
sf.k_feature_names_
```

```python
sf.k_score_
```

```python
8-0.7682624564977505
```
