```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"/Users/princegupta/Documents/data science/dataset/placement.csv")
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingRegressor
```

```python
from sklearn.svm import SVR
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
bgr=BaggingRegressor(estimator=SVR(),n_estimators=50)
bgr.fit(x_train,y_train)
```

```python
bgr.score(x_train,y_train)*100,bgr.score(x_test,y_test)*100
```

```python
rfg=RandomForestRegressor(n_estimators=50)
rfg.fit(x_train,y_train)
```

```python
rfg.score(x_train,y_train)*100,rfg.score(x_test,y_test)*100
```
