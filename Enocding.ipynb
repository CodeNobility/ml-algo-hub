# One Hot Encoding

```python
pd.get_dummies(dataset,columns=["gender"])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ar=ohe.fit_transform(dataset[["gender"]]).toarray()
```

```python
dtu=pd.DataFrame(ar,columns=ohe.get_feature_names_out(['gender']))
dtu
```

```python
dataset = pd.concat([dataset,dtu],axis=1)
```

```python
dataset
```

```python
dataset.head(2)
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dataset['Invest_Monitor'].unique()
```

```python
la=LabelEncoder()
la.fit(dataset['Invest_Monitor'])
```

```python
dataset['Invest_Monitor']=la.transform(dataset['Invest_Monitor'])
```

# Ordinal Encoding

```python
dataset1=pd.read_csv("dataset/Original_data.csv")
```

```python
dataset1.head(3)
```

```python
dataset1['How often do you monitor your investment?'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
en_data_order=[['Daily', 'Weekly', 'Monthly']]
```

```python
oen=OrdinalEncoder(categories=en_data_order)
```

```python
dataset1['How often do you monitor your investment?']=oen.fit_transform(dataset1[['How often do you monitor your investment?']])
```
